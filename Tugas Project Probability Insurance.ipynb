```python
#Pengujian Hipotesis
#1. Tagihan kesehatan perokok lebih tinggi daripada tagihan kesehatan non perokok
#2. Tagihan kesehatan dengan BMI diatas 25 lebih tinggi daripada tagihan kesehatan dengan BMI dibawah 25
```

```python
import scipy.stats as stats
```

```python
Langkah-langkah uji hipotesis:

    Menyatakan hipotesis null dan alternatif hipotesis
    Menghitung uji statistik
    Menentukan aturan keputusan
    Menyatakan rejection decision
    Kesimpulan
```

```python
import scipy.stats as stats
from sklearn.preprocessing import LabelEncoder
```

```python
#1. Tagihan kesehatan perokok lebih tinggi daripada tagihan kesehatan non perokok
Hypotesis :
    𝐻0:Tagihan Kesehatan Perokok kurang dari atau sama dengan tagihan kesehatan non perokok
    𝐻1:Tagihan Kesehatan Perokok lebih tinggi atau sama dengan tagihan kesehatan non perokok
```

```python
Alpha = 0.05
```

```python
x = df[df['smoker'] == ('yes')]['charges']
y = df[df['smoker'] == ('no')]['charges']

# performing an independent t-test
t, p_value  = stats.ttest_ind(x,y, axis = 0)  

print(p_value)
```

```python
print('Pengambilan keputusan') 
    
#Jika P_value Lebih besar dari alpha
if p_value > Alpha: 
    print('Terima 𝐻0')
#Jika P_Value lebih kecil dari Alpha
else:
    print('tolak 𝐻0')
```

```python
#2. Tagihan kesehatan dengan BMI diatas 25 lebih tinggi daripada tagihan kesehatan dengan BMI dibawah 25
Hypotesis :
    𝐻0:Tagihan Kesehatan dengan BMI Diatas 25 tidak lebih tinggi atau sama dengan tagihan kesehatan BMI dibawah 25
    𝐻1:Tagihan kesehatan dengan BMI diatas 25 lebih tinggi daripada tagihan kesehatan dengan BMI dibawah 25
```

```python
Alpha = 0.05
```

```python
x = df[df['bmi'] > 25.00] ['charges']
y = df[df['bmi'] < 25.00] ['charges']

# performing an independent t-test
t, p_value  = stats.ttest_ind(x,y, axis = 0)  

print(p_value)
```

```python
print('Pengambilan keputusan') 
    
#Jika P_value Lebih besar dari alpha
if p_value > Alpha: 
    print('Terima 𝐻0')
#Jika P_Value lebih kecil dari Alpha
else:
    print('tolak 𝐻0')
```
